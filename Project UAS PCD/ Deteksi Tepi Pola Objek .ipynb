## Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
#202331066_M.Kahlil Ghibran
```

```python
original_image_tepi = cv2.imread("Foto Diri.jpg")
#202331066_M.Kahlil Ghibran
```

```python
original_image_tepi_rgb = cv2.cvtColor(original_image_tepi, cv2.COLOR_BGR2RGB)
gray_image = cv2.cvtColor(original_image_tepi, cv2.COLOR_BGR2GRAY)
#202331066_M.Kahlil Ghibran
```

```python
canny_edges = cv2.Canny(image=gray_image, threshold1=100, threshold2=200) 
contours, hierarchy = cv2.findContours(image=canny_edges, mode=cv2.RETR_TREE, method=cv2.CHAIN_APPROX_SIMPLE)

contours_image = original_image_tepi_rgb.copy()
cv2.drawContours(image=contours_image, contours=contours, contourIdx=-1, color=(0, 255, 0), thickness=2)

print("Proses Canny Edge dan Contours Detection selesai.")
#202331066_M.Kahlil Ghibran
```

```python
fig, axs = plt.subplots(1,2, figsize = (5,5))
ax = axs.ravel()
plt.suptitle('Output 1', fontsize=16)

axs[0].imshow(original_image_tepi_rgb)
axs[0].set_title("Original Image")

axs[1].imshow(canny_edges, cmap = "gray")
axs[1].set_title("Canny Edge Detection")

plt.show()
#202331066_M.Kahlil Ghibran
```

```python
fig, axs = plt.subplots(1,3, figsize = (10,10))
ax = axs.ravel()
plt.suptitle('Output 2', fontsize=16)

axs[0].imshow(original_image_tepi_rgb)
axs[0].set_title("Original Image")

axs[1].imshow(canny_edges, cmap = "gray")
axs[1].set_title("Canny Edge Detection")

axs[2].imshow(contours_image)
axs[2].set_title("Contours Detection")

plt.show()
#202331066_M.Kahlil Ghibran
```
