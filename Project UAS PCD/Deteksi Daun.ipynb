## Import Library

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
import os
#202331066_M.Kahlil Ghibran
```

## Membaca Gambar

```python
nanas = cv2.imread("Nanas.png")
nanas_asli = cv2.imread("Foto Nanas.jpg")
#202331066_M.Kahlil Ghibran
```

```python
nanas_asli_rgb = cv2.cvtColor(nanas_asli, cv2.COLOR_BGR2RGB)

plt.imshow(nanas_asli_rgb)
plt.title("Gambar Asli Nanas Sebelum di Edit")
plt.axis('off')
plt.show()
#202331066_M.Kahlil Ghibran
```

```python
nanas_rgb = cv2.cvtColor(nanas, cv2.COLOR_BGR2RGB)

plt.imshow(nanas_rgb)
plt.title("Gambar Asli Nanas")
plt.axis('off')
plt.show()
#202331066_M.Kahlil Ghibran
```

```python
hsv_image = cv2.cvtColor(nanas, cv2.COLOR_BGR2HSV)
#202331066_M.Kahlil Ghibran
```

```python
#Membuat Mask untuk Buah Nanas
lower_buah = np.array([10, 100, 50]) 
upper_buah = np.array([25, 255, 255])

mask_buah = cv2.inRange(hsv_image, lower_buah, upper_buah)

#Membuat Mask untuk Daun
lower_daun = np.array([35, 40, 40])
upper_daun = np.array([85, 255, 255])

mask_daun = cv2.inRange(hsv_image, lower_daun, upper_daun)

print("Mask untuk buah dan daun berhasil dibuat.")
#202331066_M.Kahlil Ghibran
```

```python
segmentasi_buah = cv2.bitwise_and(nanas_rgb, nanas_rgb, mask=mask_buah)
segmentasi_daun = cv2.bitwise_and(nanas_rgb, nanas_rgb, mask=mask_daun)
#202331066_M.Kahlil Ghibran
```

```python
fig, axs = plt.subplots(2,2, figsize = (10,10))
ax = axs.ravel()
plt.suptitle("Hasil Deteksi Daun dan Buah Nanas", fontsize=16)

axs[0,0].imshow(nanas_rgb)
axs[0,0].set_title("1. Gambar Asli")

axs[0,1].imshow(mask_buah, cmap='gray')
axs[0,1].set_title("2. Mask Buah (Nanas)")

axs[1,0].imshow(segmentasi_buah)
axs[1,0].set_title("3. Segmentasi Buah (Nanas)")

axs[1,1].imshow(segmentasi_daun)
axs[1,1].set_title("4. Segmentasi Daun")

plt.show()
#202331066_M.Kahlil Ghibran
```
